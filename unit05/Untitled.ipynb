## Exe 01
题目：（统计正数和负数的个数然后计算出这些数的平均值）编写一个程序来读入不定个数的整数，然后决定已经读取的整数中有多少个正数和负数，以及它们的和与平均值。

```python
total = 0
n = 0
n_p = 0
n_n = 0

while True:
    try:
        num = int(input("Enter an integer, the input ends if it is 0: "))
        if num == 0:
            break
        total += num
        n += 1
        if num > 0: n_p += 1
        else:       n_n += 1
    except ValueError:
        break
        
if n == 0:
    print("You didn't enter any number")
else:
    average = total / n
    print("The number of positives is {}".format(n_p))
    print("The number of negetives is {}".format(n_n))
    print("The total is {}".format(total))
    print("The average is {}".format(average))
```

## Exe 02
题目：（加法）程序清单 5-4 中产生五个随机减法的问题。改写这个程序，产生两个 1 到 15 之间的随机数减法问题，显示回答正确的次数和测试使用的时间。

```python
import random
import time

correctCnt = 0
NUM_OF_QUESTION = 5

start_time = time.time()

for i in range(NUM_OF_QUESTION):
    n1 = random.randint(1, 15)
    n2 = random.randint(1, 15)
    
    ans = int(input("What is {} + {}? ".format(n1, n2)))
    if ans == n1 + n2:
        print("You'r correct!")
        correctCnt += 1
    else:
        print("You'r wrong~")

end_time = time.time()
lapse = int(end_time - start_time)

print("Correct count is {} out of {}\nTest time is {} seconds".format(correctCnt, NUM_OF_QUESTION, lapse))
```

## Exe 07
题目：（使用三角函数）打印表格显示 0 到 360 度每隔10度的sin与cos值，保留小数点后四位。

```python
import math
MAX_VAL = 360
print(34 * "=")
print("|{:^10s}|{:^10s}|{:^10s}|".format("degrees", "Sin", "Cos"))
print(34 * "=")

for i in range(0, MAX_VAL+1, 10):
    rad = math.radians(i)
    sin = math.sin(rad)
    cos = math.cos(rad)
    print("|{:^10d}|{:^10.4f}|{:^10.4f}|".format(i, sin, cos))
    if i == MAX_VAL:
        print(34 * "=")
    else:
        print(34 * "-")
```

## Exe 09
题目：（财务应用程序：计算未来的学费）假设今年的学费是 10,000 美元，且以每年 5% 增长。编写一个程序计算十年之内的每一年学费以及大学四年的总学费。

```python
def fee(year):
    return 10000 * 1.05 ** (year - 1)

def totalFee(year):
    total = 0
    for i in range(year, year + 4):
        total += fee(i)
    return total

print("{:<10s}{:<10s}{:<10s}".format("year", "fee", "4 year"))
for i in range(1, 11):
    print("{:>5d}  {:<10.2f}{:<10.2f}".format(i, fee(i), totalFee(i)))
```

## Exe 16
题目：（计算最大公约数）对于程序 5-8，另外找出两个整数 n1 和 n2 的最大公约数的解决方案为：首先找出 n1 和 n2 的最小数 d，然后以 d-1、d-2、d-3、……、2、1的顺序依次检测它们是否是 n1 和 n2 的公因子。第一个这样的公约数就是 n1 和 n2 的最大公约数

```python
n1, n2 = eval(input("Ener two number n1 and n2: "))

d = n1 if n1 < n2 else n2
gcd = 1

for i in range(d, 0, -1):
    if n1 % i == 0 and n2 % i == 0:
        gcd = i
        break
print("The great common divisor of {} and {} is {}".format(n1, n2, gcd))
```

```python
n = int(input("Enter an integer: "))

x = 2
while x <= n:
    if n % x == 0:
        print(x, end = ' ')
        n = n / x
    else:
        x += 1
```

```python
n = int(input("Enter the number of lines: "))
for i in range(1, n+1):
    for j in range(n-i, 0, -1):
        print(" "*3, end = "")
    for j in range(i, 0, -1):
        print("{:3d}".format(j), end = "")
    for j in range(2, i+1):
        print("{:3d}".format(j), end = "")
    print()
```

## Exe 20
题目：使用四种循环模式

```python
n = int(input("Enter the number of lines: "))

print("--- mode A ---")
for i in range(1, n+1):
    for j in range(1, i+1):
        print("{:<3d}".format(j), end = "")
    print()
    
print("\n--- mode B ---")
for i in range(n, 0, -1):
    for j in range(1, i+1):
        print("{:<3d}".format(j), end = "")
    print()

print("\n--- mode C ---")
for i in range(1, n+1):
    for j in range(n-i, 0, -1):
        print(" "*3, end = "")
    for j in range(i, 0, -1):
        print("{:<3d}".format(j), end = "")
    print()
    
print("\n--- mode D ---")
for i in range(n, 0, -1):
    for j in range(n-i, 0, -1):
        print(" "*3, end = "")
    for j in range(i, 0, -1):
        print("{:<3d}".format(j), end = "")
    print()
```

## Exe 22
题目：显示 2 到 1000 之间的素数，每行显示八个

```python
def isPrime(num):
    upper = int(num / 2) + 1
    for divisor in range(2, upper):
        if num % divisor == 0:
            return False
    return True
cnt = 0
for num in range(2, 1001):
    if isPrime(num):
        print("{:4d}".format(num), end=" ")
        cnt += 1
        if cnt % 8 == 0:
            print()
```
